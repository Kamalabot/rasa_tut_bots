Notebook has supporting code for econobot. The Countries_economies.csv is explored and the code is used in the action server of the bot

```python
import pandas as pd

econ_df = pd.read_csv("Countries_Economies.csv")
```

```python
# How many countries are covered

countries = econ_df.Country.unique()
len(countries)
```

```python
# What are the dimensions covered?
# Reducing / rewriting the names to make it easily readable

proc_columns = ['Country',
       'Poverty headcount ratio',
       'Life expectancy in years', 'Population',
       'Population growth in Pc', 'Net migration',
       'Human Capital Index',
       'GDP LCU',
       'GDP per capita LCU',
       'GDP growth in Pc',
       'Unemployment Pc',
       'Inflation Pc',
       'Personal remittances Pc',
       'CO2 emissions Tons',
       'Forest area Pc Land Area',
       'Access to electricity Pc',
       'Annual freshwater withdrawals Pc',
       'Electricity production Pc',
       'Sanitation services Pc of population',
       'Intentional homicides',
       'Central government debt Pc GDP',
       'SPI',
       'Individuals with Internet Pc Population',
       'Seats_held_by_women_in national parliaments Pc',
       'FDI_Pc_of_GDP', 'Unnamed']

proc_columns = [col.replace(' ','_') for col in proc_columns]
proc_columns
```

```python
econ_df.columns = proc_columns
```

```python
econ_df.head()
```

```python
econ_df[econ_df.Country == 'india']['Population'].to_list()
```

```python
#Accessing particular column data for a country

def data_access(country, column):
    if country not in econ_df.Country.unique():
        return 'Country not in list'
    
    if column not in econ_df.columns:
        return 'Column not in list for requested country'
        
    return econ_df[econ_df.Country == country][column].to_list()[0]
```

```python
data_access('India','Human_Capital_Index')
```

```python
econ_df.loc[econ_df.Population == econ_df['Population'].min(),'Country'].to_list()
```

```python
# Access the country lowest in a dimension
def lowest_in(column):
    "Which country is minimum in a dimension"
    
    if column not in econ_df.columns:
        return 'Column requested not in list'
        
    minimum = econ_df[column].min()    
    
    return econ_df.loc[econ_df[column] == minimum,'Country'].to_list()[0]
```

```python
lowest_in('Population')
```

```python
# Access the country highest in a dimension
def highest_in(column):
    "Which country is maximum in a dimension"
    
    if column not in econ_df.columns:
        return 'Column requested not in list'
        
    maximum = econ_df[column].max()    
    
    return econ_df.loc[econ_df[column] == maximum,'Country'].to_list()[0]
```

```python
highest_in('Population')
```
